## Itération 4

```python
class CompteurDesMots:
    
    def __init__(self) -> None:
        self._frequence_mots = {}

    def compter_mots(self, texte) -> None:
        """
        Compte les mots dans un texte et mets à jour 
        le dictionnaire interne avec la fréquence des mots.
        """
        for mot in texte.split(' '):
            if mot not in self._frequence_mots:
                self._frequence_mots[mot] = 1
            else:
                self._frequence_mots[mot] += 1

    def obtenir_fréquence(self, mot) -> int:
        """
        Si le mot est présent, sa fréquence est retournée, 
        sinon un message d'erreur est affiché.
        """
        if mot in self._frequence_mots:
            return self._frequence_mots[mot]
        else:
            print(f"le mot {mot} n'est pas présent")
```

```python
mon_compteur_de_mots = CompteurDesMots()
mon_compteur_de_mots.compter_mots(contenu)
mon_compteur_de_mots.obtenir_fréquence('δὲ')
```

```python
mon_compteur_de_mots.obtenir_fréquence('τα')
```
